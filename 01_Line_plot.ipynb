# to LinePlot hamesha numerical var ke sath banta h. isme X or Y axis pr continous var liye jaate hn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
phool = sns.load_dataset('iris')
```

```python
phool.head()
```

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
plt.show()
```

### adding Title

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
plt.title("phoolo ka plot")
plt.show()
```

### Adding limit

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
plt.title("phoolo ka plot")
plt.xlim(2)
plt.ylim(1)
plt.show()
```

### Set Style

- darkgrid
- whitegrid
- dark
- white
- ticks

```python
sns.set_style(style=None, rc=None)
```

```python
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
plt.title("phoolo ka plot")
sns.set_style("dark")
plt.show()
```

# Figure Size

```python
plt.figure(figsize=(8,6))
sns.lineplot(x='sepal_length', y='sepal_width', data=phool)
plt.title('phoolo ka plot')
plt.show()
```
